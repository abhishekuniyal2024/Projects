# BREAST CANCER DETECTION USNG RANDOM FOREST TREE

```python
# Breast cancer detection using RANDOM FOREST TREE 

# import library
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier 
from sklearn.metrics import confusion_matrix, accuracy_score
import numpy as np

# load data
data = pd.read_csv("E:/Ducat/Machine Learning/Breast_cancer.csv")

# Split data into features (X) and target (y)
X, y = data.iloc[:, :-1], data.iloc[:, -1]

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Train Random Forest model
random_forest = RandomForestClassifier(random_state=0)  # Initialize the RandomForestClassifier
random_forest.fit(x_train, y_train)  # Train the model

# Make predictions
y_pred = random_forest.predict(x_test)  # Predict the classes for the test set

# Evaluate the model
result = confusion_matrix(y_test, y_pred)  # Get the confusion matrix
print('Confusion Matrix:')
print(result)
print('Accuracy Score:', accuracy_score(y_test, y_pred))  # Print the accuracy score

# Predict for a single sample
k = np.array([[15, 10, 5, 100, 0.1]])

# Convert k to a DataFrame with the same columns as X
k_df = pd.DataFrame(k, columns=X.columns)  # Ensure k has the same feature names as X

# Predict the class for k using the DataFrame
pred = random_forest.predict(k_df)  # Predict the class for the input data
print("Prediction for Breast Cancer:", pred)  # Print the prediction for the sample
```

# Breast Cancer Detection using Random Forest

This project uses the **Random Forest** algorithm to detect breast cancer. Below are the steps involved in the project:

### Steps Involved:
1. **Importing Libraries**:  
   Essential libraries such as `pandas`, `sklearn`, and `numpy` are imported for data handling, machine learning, and evaluation.

2. **Loading the Dataset**:  
   The dataset is loaded into a Pandas DataFrame from a CSV file. It contains features and a target column indicating cancer diagnosis (e.g., `0` for benign, `1` for malignant).

3. **Data Splitting**:  
   - The data is split into features (`X`) and the target (`y`).  
   - Further split into training and testing sets using `train_test_split`, with 30% of data allocated for testing.

4. **Training the Random Forest Model**:  
   - A `RandomForestClassifier` is initialized and trained on the training set (`x_train`, `y_train`).  
   - The trained model is used to make predictions on the test set (`x_test`).

5. **Model Evaluation**:  
   - The performance is evaluated using a **confusion matrix** and the **accuracy score**.  
   - The confusion matrix provides details of correct and incorrect predictions, while the accuracy score represents overall model accuracy.

6. **Single Sample Prediction**:  
   - A sample input is created as a NumPy array and converted to a DataFrame with the same feature names as the dataset.  
   - The trained model predicts the class (benign or malignant) for this sample input.
